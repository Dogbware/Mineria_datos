```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('amazon1.csv', on_bad_lines='skip', encoding='latin1', engine='python')
```

```python
df.info()
```

```python
df.describe()
```

```python
df_nuevo = df.dropna(axis=1, how='all')
```

```python
df_nuevo.fillna(0, inplace=True)
df_nuevo
```

```python
numeric_df = df_nuevo.select_dtypes(include=['number'])
numeric_df.describe()
```

```python
matriz=numeric_df.corr(method="pearson")
matriz
```

```python
plt.figure(figsize=(10,6))
plt.imshow(matriz, cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(label="relacion")
for i in range(len(matriz)):
  for j in range(len(matriz)):
    plt.text(j, i, f"{matriz.iloc[i, j]:.2f}", ha="center", va="center", color="black")
plt.xticks(range(len(matriz.columns)), matriz.columns, rotation=45)
plt.yticks(range(len(matriz.index)), matriz.index)
plt.title('matriz de correlacion')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

data = {
    "Horas_estudio": [1,2,1,5,8,9,8.5,3,3,5,4],
    "Calificacion": [2,3,2,5,9,10,9.5,5,5.5,6,8]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["Horas_estudio"], df["Calificacion"])
plt.xlabel("Horas de estudio")
plt.ylabel("Calificacion")
plt.title("Datos de los estudiantes")
plt

# 1. Crear una lista para guardar las inercias
inercias = []

# 2. Probar con diferentes números de clústeres (por ejemplo, del 1 al 10)
for k in range(1, 11):
    kmeans_prueba = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans_prueba.fit(df[["Horas_estudio", "Calificacion"]])
    inercias.append(kmeans_prueba.inertia_) # Guardamos la inercia

# 3. Graficar el "Codo"
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inercias, marker='o', linestyle='--', color='b')
plt.title('Método del Codo')
plt.xlabel('Número de clústeres (k)')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()

kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df[["Horas_estudio", "Calificacion"]])
kmeans.fit(df)

centroides = kmeans.cluster_centers_
centroides

df["Cluster"] = kmeans.labels_
df

plt.scatter(df["Horas_estudio"], df["Calificacion"], c=df["Cluster"])
plt.scatter(centroides[:,0], centroides[:,1], marker="x", color="red", s=200, label="centroides")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificacion")
plt
```

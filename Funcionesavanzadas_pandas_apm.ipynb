```python
import pandas as pd
df = pd.read_csv("empleados.csv")
```

```python
resultado = df.loc[
    (df["dept"] == "Eng") & (df["perf_score"] >= 0.85),
    ["emp_id", "name", "city", "perf_score", "salary_mxn"]
]

print(resultado)
```

```python
resultado = df.iloc[3:8, [0, 1, 5, 6]]

print(resultado)
```

```python
resultado = df.loc[
    (df["city"].isin(["CDMX", "GDL"])) &
    (df["warnings"] == 0) &
    (df["absences"] <= 1),
    ["emp_id", "name", "city", "warnings", "absences"]
]

print(resultado)
```

```python
resumen = df.agg(
    salario_promedio=("salary_mxn", "mean"),
    perf_promedio=("perf_score", "mean"),
    total_warnings=("warnings", "sum")
)

print(resumen)
```

```python
resumen_depto = (
    df.groupby("dept")
      .agg(
          num_empleados=("emp_id", "count"),
          salario_promedio=("salary_mxn", "mean"),
          perf_promedio=("perf_score", "mean")
      )
      .sort_values("salario_promedio", ascending=False)
)

print(resumen_depto)
```
